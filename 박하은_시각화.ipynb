```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import os

# 데이터를 사이트별로 분리하기 위한 함수
def split_by_site(df):
    sites = {}
    for index, row in df.iterrows():
        site = row['url'].split('.onion')[0]
        if site not in sites:
            sites[site] = []
        sites[site].append(row['product'])
    return sites

# 워드클라우드 생성 함수 (지정한 단어 제외)
# 워드클라우드 생성 함수 (지정한 단어 제외)
def generate_wordcloud(data, excluded_words, save_dir):
    for i, (site, products) in enumerate(data.items(), 1):
        site_name = site.replace(':', '_').replace('/', '_').replace('.', '_')
        text = ' '.join([word for word in products if not any(excluded_word in word for excluded_word in excluded_words)])
        wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.title(f"Word Cloud for Site: {site}")
        plt.axis('off')
        # Save the word cloud as an image with numerical filename
        save_path = os.path.join(save_dir, f"{i}_wordcloud.png")
        plt.savefig(save_path)
        print(f"Word cloud saved as: {save_path}")
        plt.close()

# URL에서 ".onion"까지를 기준으로 데이터 분리
sites_data = split_by_site(df)

# 제외할 단어 목록
excluded_words = ["USA","online","M99","5G","280MG","TOP SHELF","pack","1G","2MG","GRAM","Ready","gram"]

# 저장할 디렉토리 설정
save_dir = '/content/gdrive/MyDrive/save_path'

# 각 사이트별로 워드클라우드 생성 (제외할 단어 적용)
generate_wordcloud(sites_data, excluded_words, save_dir)

```
